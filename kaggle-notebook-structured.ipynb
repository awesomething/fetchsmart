````python
# Setup and Installation
# This cell installs required dependencies for the multi-agent system

import sys
import subprocess

def install_package(package):
    """Install a Python package if not already installed"""
    try:
        __import__(package)
        print(f"âœ… {package} is already installed")
    except ImportError:
        print(f"ğŸ“¦ Installing {package}...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", package, "-q"])
        print(f"âœ… {package} installed successfully")

# Core dependencies
packages = [
    "matplotlib",  # For architecture visualization
    "numpy",  # For visualization support
]

print("ğŸ”§ Setting up environment for multi-agent system...\n")
for package in packages:
    install_package(package)

print("\nâœ… Setup complete! Ready to explore multi-agent systems.")
````

````python
# Sequential Agent Pipeline Demonstration
# Shows how the 4-stage email generation pipeline works

class SequentialPipeline:
    """Demonstrates the sequential agent pipeline pattern"""
    
    def __init__(self):
        self.stages = []
        self.results = {}
    
    def add_stage(self, name, agent, description):
        """Add a stage to the pipeline"""
        self.stages.append({
            'name': name,
            'agent': agent,
            'description': description
        })
    
    def execute(self, input_data):
        """Simulate pipeline execution"""
        print(f"\nğŸš€ Starting Sequential Pipeline with input: {input_data[:50]}...\n")
        
        current_data = input_data
        for i, stage in enumerate(self.stages, 1):
            print(f"Stage {i}: {stage['name']}")
            print(f"   Agent: {stage['agent']}")
            print(f"   Action: {stage['description']}")
            
            # Simulate processing
            current_data = f"{current_data} â†’ [Processed by {stage['name']}]"
            self.results[stage['name']] = current_data
            
            print(f"   âœ… Complete\n")
        
        return current_data

# Create email generation pipeline
email_pipeline = SequentialPipeline()

email_pipeline.add_stage(
    "Email Generator",
    "EmailGenerator",
    "Creates initial email draft (75-125 words)"
)

email_pipeline.add_stage(
    "Email Reviewer",
    "EmailReviewer",
    "Validates email quality, tone, and length"
)

email_pipeline.add_stage(
    "Email Refiner",
    "EmailRefiner",
    "Improves email based on feedback (optional)"
)

email_pipeline.add_stage(
    "Email Presenter",
    "EmailPresenter",
    "Formats final email with Next Steps section"
)

# Demonstrate pipeline execution
input_data = "Job: Senior React Developer | Candidate: John Doe (GitHub: johndoe)"
result = email_pipeline.execute(input_data)

print("=" * 60)
print("PIPELINE EXECUTION COMPLETE")
print("=" * 60)
print(f"\nFinal Output: {result}")
print("\nğŸ“Š Pipeline Characteristics:")
print("   - Sequential execution (each stage depends on previous)")
print("   - State passed between agents")
print("   - Optional refinement stage")
print("   - Full observability at each stage")
print("=" * 60)
````

````python
# Smart Routing Demonstration
# Shows how the root agent routes queries to specialized agents

class SmartRouter:
    """Demonstrates intelligent agent routing"""
    
    def __init__(self):
        self.agents = {
            "planning": "HiringPlanAgent",
            "faq": "FAQAgent",
            "recruiter": "RecruiterOrchestrator",
            "email": "RecruiterEmailPipeline",
            "staffing_recruiter": "StaffingRecruiterOrchestrator",
            "staffing_employer": "StaffingEmployerOrchestrator"
        }
    
    def route(self, query):
        """Route query to appropriate agent"""
        query_lower = query.lower()
        
        # Routing logic
        if any(word in query_lower for word in ["plan", "schedule", "task"]):
            return self.agents["planning"]
        elif any(word in query_lower for word in ["what", "how", "question", "faq"]):
            return self.agents["faq"]
        elif any(word in query_lower for word in ["find", "search", "candidate", "engineer"]):
            return self.agents["recruiter"]
        elif any(word in query_lower for word in ["email", "generate", "outreach"]):
            return self.agents["email"]
        elif any(word in query_lower for word in ["job", "position", "opening"]):
            return self.agents["staffing_recruiter"]
        elif any(word in query_lower for word in ["review", "evaluate", "candidate"]):
            return self.agents["staffing_employer"]
        else:
            return "RootAgent"  # Default to root for complex queries
    
    def demonstrate(self, queries):
        """Demonstrate routing for multiple queries"""
        print("=" * 60)
        print("SMART ROUTING DEMONSTRATION")
        print("=" * 60)
        
        for query in queries:
            agent = self.route(query)
            print(f"\nğŸ“ Query: \"{query}\"")
            print(f"   â†’ Routes to: {agent}")
        
        print("\n" + "=" * 60)
        print("âœ… Smart Routing Features:")
        print("   - Intent detection from query text")
        print("   - Automatic agent selection")
        print("   - Explicit mode overrides supported ([MODE:XXX])")
        print("   - Fallback to root agent for complex queries")
        print("=" * 60)

# Create router and demonstrate
router = SmartRouter()

test_queries = [
    "Find 10 senior React engineers",
    "Generate outreach email for candidate John Doe",
    "Plan my week for filling 3 positions",
    "What is our deployment process?",
    "Search for job openings in San Francisco"
]

router.demonstrate(test_queries)
````

````python
# Weave Tracing Demonstration
# Shows how observability works in the multi-agent system

class TraceSpan:
    """Represents a trace span for observability"""
    
    def __init__(self, name, service, operation, duration_ms=0):
        self.name = name
        self.service = service
        self.operation = operation
        self.duration_ms = duration_ms
        self.children = []
    
    def add_child(self, child_span):
        """Add a child span"""
        self.children.append(child_span)
    
    def display(self, indent=0):
        """Display span hierarchy"""
        prefix = "  " * indent
        print(f"{prefix}ğŸ“Š {self.name}")
        print(f"{prefix}   Service: {self.service}")
        print(f"{prefix}   Operation: {self.operation}")
        print(f"{prefix}   Duration: {self.duration_ms}ms")
        
        for child in self.children:
            child.display(indent + 1)

# Create example trace
root_trace = TraceSpan(
    name="User Query: Find React Engineers",
    service="adk-agent",
    operation="root_query",
    duration_ms=2500
)

# Agent span
agent_span = TraceSpan(
    name="RecruiterOrchestrator",
    service="adk-agent",
    operation="candidate_search",
    duration_ms=1800
)

# MCP tool spans
tool1_span = TraceSpan(
    name="search_candidates_tool",
    service="recruitment-backend",
    operation="mcp_tool_call",
    duration_ms=1200
)

tool2_span = TraceSpan(
    name="scrape_github_profiles_tool",
    service="recruitment-backend",
    operation="mcp_tool_call",
    duration_ms=800
)

# Build hierarchy
agent_span.add_child(tool1_span)
agent_span.add_child(tool2_span)
root_trace.add_child(agent_span)

# Display trace
print("=" * 60)
print("WEAVE TRACING DEMONSTRATION")
print("=" * 60)
print("\nğŸ“ˆ Distributed Trace Hierarchy:\n")
root_trace.display()

print("\n" + "=" * 60)
print("âœ… Observability Features:")
print("   - Full trace hierarchy (root â†’ agent â†’ tools)")
print("   - Duration tracking for performance analysis")
print("   - Service identification for distributed systems")
print("   - MCP tool call visibility")
print("   - Export to Weave (W&B) for visualization")
print("=" * 60)
````
